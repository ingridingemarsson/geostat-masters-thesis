```python
import numpy as np
import os
from pathlib import Path
```

```python
def extractSingles(path_to_boxes, path_to_singles, num_channels=8):
    if not Path.exists(Path(path_to_singles)):
        os.makedirs(path_to_singles)
        
    arr = np.empty((0,num_channels+1))
    i=0
    for f in os.listdir(path_to_boxes):
            box = np.load(os.path.join(path_to_boxes,f))
            box = box.reshape(box.shape[0], -1).T
            mask = (np.isnan(box[:,-1]) | np.less(box[:,-1], 0))
            box = box[~mask, :]
            arr = np.append(arr, box, axis=0)
            i+=1
            if i%50==0:
                print(i)
            #print(arr.shape)
            #print(arr[:,:-1].shape)
    np.save(os.path.join(path_to_singles, 'X.npy'), arr[:,:-1])
    np.save(os.path.join(path_to_singles, 'y.npy'), arr[:,-1])
```

```python
f = open('../../path_to_data.txt', "r")
path_to_data = os.path.join(f.read(), 'data')
f.close() 
subfolder = 'npy_files/'
```

```python
extractSingles(os.path.join(path_to_data, 'dataset-boxes', 'train', subfolder), os.path.join(path_to_data, 'dataset-singles', 'train'))
```

```python
extractSingles(os.path.join(path_to_data, 'dataset-boxes', 'validation', subfolder), os.path.join(path_to_data, 'dataset-singles', 'validation'))
```

```python
extractSingles(os.path.join(path_to_data, 'dataset-boxes', 'test', subfolder), os.path.join(path_to_data, 'dataset-singles', 'test'))
```
